```python
def _plot(img1, img2):
    plt.figure(figsize=(15,8))
    plt.subplot(1, 2, 1)
    plt.imshow(cv.cvtColor(img1, cv.COLOR_BGR2RGB))
    plt.title("source_1")
    plt.subplot(1, 2, 2)
    plt.imshow(cv.cvtColor(img2, cv.COLOR_BGR2RGB))
    plt.title("source_2");
```

```python
def detector_windowed(in_fname_1, in_fname_2, out_fname=None):

    # Считываем входящие данные
    arr1 = np.array(in_fname_1)[:,:,0:3]
    arr2 = np.array(in_fname_2)[:,:,0:3]
    shape = in_fname_2.shape

    # Создаем переменные детектирования
    detections = np.zeros((shape[0], shape[1]), dtype='uint8')

    # Задаем параметры бегущего окна
    step = 20
    win = 100
    for i in range(0, shape[0]-win, step):
        for j in range(0, shape[1]-win, step):
            chip1 = arr1[i:i+win,j:j+win,:] 
            chip2 = arr2[i:i+win,j:j+win,:]
            hist_1 = cv.calcHist([cv.cvtColor(chip1.copy(), cv.COLOR_RGB2GRAY)], [0], None, [256], [0,256])
            hist_2 = cv.calcHist([cv.cvtColor(chip2.copy(), cv.COLOR_RGB2GRAY)], [0], None, [256], [0,256])
            cor_ge = cv.compareHist(hist_1, hist_2, cv.HISTCMP_CORREL)
            #print(cor_ge)
            res = compare( phash(chip1), phash(chip2))
            if cor_ge <= 0.8 and res == 1:
                detections[i:i+step,j:j+step] = 125
                #print(detections[i:i+win,j:j+win])
           
    return detections
```

```python
def compare(hesh1, hesh2):
    dif_num = 0
    for index in range(len(hesh1)):
        if hesh1[index] != hesh2[index]:
            dif_num +=1
    if dif_num > 3:
        return 1
    else:
        return 0
```

```python
def phash(chip):
    chip = cv.cvtColor(chip, cv.COLOR_BGR2GRAY)
    chip = cv.resize(chip,(32, 32))
    imf = np.float32(chip)/255.0
    chip = cv.dct(imf)
    chip = chip[0:8,0:8]
    average_color = 0
    for i in range(0, 8):
        for j in range(0, 8): 
            average_color += chip[i][j]
    average_color /= 63
    hesh = ''
    for i in range(0, 8):
        for j in range(0, 8): 
            if chip[i][j] >= average_color:
                hesh += '1'
            else:
                hesh += '0'
    hesh = hex(int(hesh, 2))
    mod = ''
    for i in range(0, 18 - len(hesh)):
        mod += '0'
    hesh = mod + hesh
    
    return hesh
```

```python
def erosion_correction(dif, kernel_size=3):
    kernel = np.ones((kernel_size, kernel_size),np.uint8)
    opening = cv.morphologyEx(dif, cv.MORPH_OPEN, kernel)
    opening = cv.morphologyEx(opening, cv.MORPH_CLOSE, kernel)
    return opening
```

```python
def rle_encode(img):
    '''
    img: numpy array, 1 - mask, 0 - background
    Returns run length as string formated
    '''
    pixels = img.T.flatten()
    pixels = np.concatenate([[0], pixels, [0]])
    runs = np.where(pixels[1:] != pixels[:-1])[0] + 1
    runs[1::2] -= runs[::2]
    return ' '.join(str(x) for x in runs)
```

```python
def rle_decode(mask_rle, shape=(768, 768)):
    '''
    mask_rle: run-length as string formated (start length)
    shape: (height,width) of array to return 
    Returns numpy array, 1 - mask, 0 - background
    '''
    s = mask_rle.split()
    starts, lengths = [np.asarray(x, dtype=int) for x in (s[0:][::2], s[1:][::2])]
    starts -= 1
    ends = starts + lengths
    img = np.zeros(shape[0]*shape[1], dtype=np.uint8)
    for lo, hi in zip(starts, ends):
        img[lo:hi] = 1
    return img.reshape(shape).T  # Needed to align to RLE direction
```

```python
import cv2
import numpy
from matplotlib import pyplot as plt

opencv_haystack = cv2.imread("img/img6.jpg")
opencv_needle = cv2.imread("img/img8.jpg")

ngrey = cv2.cvtColor(opencv_needle, cv2.COLOR_BGR2GRAY)
hgrey = cv2.cvtColor(opencv_haystack, cv2.COLOR_BGR2GRAY)

# build feature detector and descriptor extractor
hessian_threshold = 5000
detector = cv2.xfeatures2d.SURF_create(hessian_threshold)
hkeypoints,hdescriptors = detector.detectAndCompute(hgrey,None)
nkeypoints,ndescriptors = detector.detectAndCompute(ngrey,None)

# extract vectors of size 64 from raw descriptors numpy arrays
rowsize = len(hdescriptors) / len(hkeypoints)
if rowsize > 1:
    hrows = numpy.array(hdescriptors, dtype = numpy.float32).reshape((-1, rowsize))
    nrows = numpy.array(ndescriptors, dtype = numpy.float32).reshape((-1, rowsize))
    #print hrows.shape, nrows.shape
else:
    hrows = numpy.array(hdescriptors, dtype = numpy.float32)
    nrows = numpy.array(ndescriptors, dtype = numpy.float32)
    rowsize = len(hrows[0])

# kNN training - learn mapping from hrow to hkeypoints index
samples = hrows
responses = numpy.arange(len(hkeypoints), dtype = numpy.float32)
#print len(samples), len(responses)
knn = cv2.ml.KNearest_create()
knn.train(samples,cv2.ml.ROW_SAMPLE, responses)

# retrieve index and value through enumeration
for i, descriptor in enumerate(nrows):
    descriptor = numpy.array(descriptor, dtype = numpy.float32).reshape((1, rowsize))
    #print i, descriptor.shape, samples[0].shape
    retval, results, neigh_resp, dists = knn.findNearest(descriptor, 1)
    res, dist =  int(results[0][0]), dists[0][0]
    #print res, dist

    if dist < 0.1:
        # draw matched keypoints in red color
        color = (0, 0, 255)
    else:
        # draw unmatched in blue color
        color = (255, 0, 0)
    # draw matched key points on haystack image
    x,y = hkeypoints[res].pt
    center = (int(x),int(y))
    cv2.circle(opencv_haystack,center,5,color,-1)
    # draw matched key points on needle image
    x,y = nkeypoints[i].pt
    center = (int(x),int(y))
    cv2.circle(opencv_needle,center,5,color,-1)

bf = cv2.BFMatcher(cv2.NORM_L1,crossCheck=True) # crossCheck not supported by BFMatcher
clusters = numpy.array([hdescriptors])
bf.add(clusters)

# Train: Does nothing for BruteForceMatcher though.
bf.train()

matches = bf.match(ndescriptors)
matches = sorted(matches, key = lambda x:x.distance)
nmatches = 10

img_matches = cv2.drawMatches(ngrey, nkeypoints, hgrey, hkeypoints, matches[:nmatches], hgrey, flags=2) # Show top 10 matches
plt.figure(figsize=(30, 30))
plt.title(type(detector))
plt.imshow(img_matches)
plt.show()   

coord_list1 = []
coord_list2 = []

for index in range(0 , nmatches):
    mat = matches[index]
    # Get the matching keypoints for each of the images
    img1_idx = mat.queryIdx
    img2_idx = mat.trainIdx
    (x1,y1) = nkeypoints[img1_idx].pt
    (x2,y2) = nkeypoints[img2_idx].pt
    coord_list1.append([x1,y1])
    coord_list2.append([x2,y2])
    
    
print(coord_list1,'\n' ,coord_list2)    
#cv2.imwrite("C:\\Users\\Dee\\Downloads\\bbc.jpg", opencv_haystack)
#cv2.imwrite("C:\\Users\\Dee\\Downloads\\bcc.jpg", opencv_needle)
#cv2.imwrite("C:\\Users\\Dee\\Downloads\\result_wo.jpg", img_matches)
```

```python
def image_detect_and_compute(detector, img_name):
    """Detect and compute interest points and their descriptors."""
    img = cv2.imread(img_name)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    kp, des = detector.detectAndCompute(img, None)
    return img, kp, des
    
def draw_image_matches(detector, img1_name, img2_name, nmatches=3):
    coord_list1 = []
    coord_list2 = []
    """Draw ORB feature matches of the given two images."""
    img1, kp1, des1 = image_detect_and_compute(detector, img1_name)
    img2, kp2, des2 = image_detect_and_compute(detector, img2_name)
    
    bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
    matches = bf.match(des1, des2)
    
    matches = sorted(matches, key = lambda x: x.distance) # Sort matches by distance.  Best come first.
    for index in range(0 , nmatches):
        mat = matches[index]
        # Get the matching keypoints for each of the images
        img1_idx = mat.queryIdx
        img2_idx = mat.trainIdx
        # x - columns
        # y - rows
        # Get the coordinates
        (x1,y1) = kp1[img1_idx].pt
        (x2,y2) = kp2[img2_idx].pt
        coord_list1.append([x1,y1])
        coord_list2.append([x2,y2])
    #img_matches = cv2.drawMatches(img1, kp1, img2, kp2, matches[:nmatches], img2, flags=2) # Show top 10 matches
    #plt.figure(figsize=(16, 16))
    #plt.title(type(detector))
    #plt.imshow(img_matches); plt.show()
    return (coord_list1, coord_list2)
```

```python

def fourier(_img):

    #img = cv2.cvtColor(_img, cv2.COLOR_BGR2GRAY)
    img = _img
    r = 100

    f = np.fft.fft2(img)
    fshift = np.fft.fftshift(f)
    magnitude_spectrum = 20*np.log(np.abs(fshift))
    M = np.max(magnitude_spectrum)
    coord = np.where(magnitude_spectrum == M)
    for x in range(0, len(fshift[0])):
        for y in range(0, len(fshift[1])):
            if ((x - coord[0])**2 + (y - coord[1])**2 ) < r**2:
                fshift[x][y] = 1
    peps = np.fft.ifftshift(fshift)
    peps = np.fft.ifft2(peps)
    peps = np.real(peps)
    return peps
    
def find_borders(img):
    gray = cv2.GaussianBlur(img, (3, 3), 0)
    slice1Copy = np.uint8(gray)
    edged = cv2.Canny(slice1Copy, 10, 100)
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (7, 7))
    closed = cv2.morphologyEx(edged, cv2.MORPH_CLOSE, kernel)
    th3 = cv2.adaptiveThreshold(closed, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 3)
    return th3

```
